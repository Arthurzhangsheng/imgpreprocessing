```python
import cv2
import os
from pathlib import PurePath, Path
from glob import glob

read_path = r"D:\jpg_data\gongzheng\512"
save_path = r"D:\jpg_data\gongzheng\256"

#创建保存文件夹
Path(f"{save_path}").mkdir(parents=True, exist_ok=True)

#读取所有图片名
all_imgs = glob(f"{read_path}/*.*")

for imgname in all_imgs:    
    print(f"正在处理{imgname}")
    raw_imgname = PurePath(imgname).parts[-1]
    img = cv2.imread(imgname)
    
    res = cv2.resize(img,(256,256),interpolation=cv2.INTER_AREA)

    #保存图片
    cv2.imwrite(f"{save_path}/{raw_imgname}",res)
```
