```python
from mtcnn.mtcnn import MTCNN
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 15)
```

```python
img_path = "imgs/multiple_faces.jpg"
img = cv2.imdecode(np.fromfile(img_path,dtype=np.uint8),-1)
detector = MTCNN()

result = detector.detect_faces(img)
print(result)
```

```python
img_rect = img.copy()
if len(result)>0:#如果人脸数组长度大于0
    for face_info in result: #对每一个人脸画矩形框
        #画脸框
        x, y, w, h = face_info["box"]
        cv2.rectangle(img_rect, (x, y), (x+w, y+h), (255,0,0), thickness=2)
        #画关键点
        for _,point in face_info["keypoints"].items():
            cv2.circle(img_rect, point, 1, (255,0,0), thickness=2)

img_rect_rgb = img_rect[...,::-1]
plt.imshow(img_rect_rgb)
plt.show()
```
